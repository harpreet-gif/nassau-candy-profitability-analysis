## DATA CLEANING

#### Convert dates

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst = True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst = True)
```

```python
df.head()
```

#### Remove invalid rows

```python
df = df[df['Sales'] > 0]
df = df[df['Gross Profit'] != 0]
```

#### Standardize division names

```python
df['Division'] = df['Division'].str.strip().str.title()
```

```python
df.head()
```

## CREATE KPI METRICS

#### Gross Margin %

##### Gross Margin % = Gross Profit/sale * 100

```python
df['Gross Margin %'] = (df['Gross Profit'] / df['Sales']) * 100
```

```python
df
```

#### Profit Per Unit

##### Profit Per Unit = Gross Profit / units

```python
df['Profit Per Unit'] = df['Gross Profit'] / df['Units']
```

```python
df
```

#### Revenue Contribution

##### Revenue Contribution = Product sales / Total sales * 100

```python
total_sales = df['Sales'].sum()

df['Revenue Contribution %'] = (
    df['Sales'] / total_sales
) * 100
```

```python
df.head()
```

#### Profit Contribution

##### Profit Contribution = product profit / total profit * 100

```python
total_profit = df['Gross Profit'].sum()

df['Profit Contribution %'] = (
    df['Gross Profit'] / total_profit
) * 100
```

```python
df.head()
```

## PRODUCT LEVEL PROFITABILITY ANALYSIS

#### Top Profitable Products

```python
top_profit_products = df.groupby('Product Name')['Gross Profit'].sum().sort_values(ascending=False)

print(top_profit_products.head(10))
```

#### Top Margin Products

```python
top_margin_products = df.groupby('Product Name')['Gross Margin %'].mean().sort_values(ascending=False)

print(top_margin_products.head(10))
```

#### High Sales but Low Margin Products

```python
high_sales_low_margin = df[
    (df['Sales'] > df['Sales'].quantile(0.75)) &
    (df['Gross Margin %'] < df['Gross Margin %'].quantile(0.25))
]

high_sales_low_margin[['Product Name', 'Sales', 'Gross Margin %']]
```

## DIVISION PERFORMANCE ANALYSIS

```python
division_analysis = df.groupby('Division').agg({
    'Sales': 'sum',
    'Gross Profit': 'sum',
    'Gross Margin %': 'mean'
}).reset_index()

print(division_analysis)
```

## PARETO ANALYSIS (80/20 RULE)

#### Revenue Pareto

```python
pareto = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)

pareto_df = pareto.reset_index()

pareto_df['Cumulative Sales'] = pareto_df['Sales'].cumsum()

pareto_df['Cumulative %'] = (
    pareto_df['Cumulative Sales'] / pareto_df['Sales'].sum()
) * 100

pareto_df.head()
```

#### Profit Pareto

```python
profit_pareto = df.groupby('Product Name')['Gross Profit'].sum().sort_values(ascending=False)

profit_df = profit_pareto.reset_index()

profit_df['Cumulative Profit'] = profit_df['Gross Profit'].cumsum()

profit_df['Cumulative %'] = (
    profit_df['Cumulative Profit'] / profit_df['Gross Profit'].sum()
) * 100
```
